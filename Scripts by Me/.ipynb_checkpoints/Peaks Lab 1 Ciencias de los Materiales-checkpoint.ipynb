# Imports Brutales

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
```

# Codigo generador

Para cada peak de cada material definiremos una función gaussiana la cual es un ajuste a los datos obtenidos por medio del programa WebPlotDigitizer.

## Material 1

1. Generamos una instancia de grados por medio de la libreria numpy desde 20 hasta 80 grados.
2. Creamos las funciones para cada peak segun los parámetros obtenidos.

```python
t=np.linspace(20,80,1000)
p = lambda t,a,h,o,w: a+h*np.e**(-(t-o)**2/(2*w**2 )) 
fp1 = p(t,41.3,821.7,38.3,0.287)
fp2 = p(t,52.4 ,552.9,46.3, 0.255)
fp3 = p(t,22.6 ,384.9,64.9, 0.277)
fp4 = p(t,16.8 ,55.1,78, 0.297)
```

```python
plt.figure(figsize=(12,5))
plt.plot(t, fp1,label='Peak 1')
plt.plot(t, fp2,label='Peak 2')
plt.plot(t, fp3,label='Peak 3')
plt.plot(t, fp4,label='Peak 4')
plt.title('Peaks del material 1')
plt.ylabel('Intensidad')
plt.xlabel('Angulo (2O)')
plt.legend(loc='best')
plt.show()
```
