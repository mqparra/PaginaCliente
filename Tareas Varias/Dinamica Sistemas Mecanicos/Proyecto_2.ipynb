Primero definimos los parametros necesarios

```python
# Parametros
R = 50 #m
M = 1900 #Kg
l = 5 #m
a = 1.8 #m
h = 1.7 #m
w = 0.6 #m
v = 20 #m/s
```

Definimos las masas para cada caso que deseamos investigar

```python
m1=M+65*4
m2=M+65*6
m3=M+65*8
```

Revisamos la condicion de volcamiento

$$\sqrt{Rg(\frac{a/2}{h/2+2w})}<v$$

```python
condicion=np.sqrt(R*g*a/(h+4*w))
if condicion <v:
    print('El automovil se vuelca ya que {:.1f} es menor que {}'.format(condicion,v))
else:
    print('El automovil no se vuelca ya que {:.1f} es mayor que {}'.format(condicion,v))
```

Lo cual es menor a v, por ende se vuelca

$$I_Q= m\frac{(a^2+h^2)}{12}+M\cdot d^2$$

$$I_Q=m\frac{(a^2+h^2)}{12}+m\frac{(a^2+(h+2w)^2)}{4}$$

Generamos un espacio lineal para los timepos desde 0 a 100 segundos por mediode la libreria numpy

```python
t=np.linspace(0,10,100)
```

```python
epsilon = lambda m: -a/2*m*g+(h+2*w)*m*v**2/R
I_zz = lambda m:m*(a**2+(h)**2)/12+ m*(a**2+(h+2*w)**2)/4
division = lambda m: epsilon(m)/I_zz(m) 
```

### Parte 4

```python
teta_dot = lambda t,m: division(m)*t
teta_t = lambda t,m: teta_dot(t,m)**2/2*division(m)**(-1)
```

```python
plt.figure(figsize=(12,7))
plt.plot(t, teta_t(t,m1),label='$θ(t)_1$')
plt.plot(t, teta_t(t,m2),label='$θ(t)_2$')
plt.plot(t, teta_t(t,m3),label='$θ(t)_3$')


plt.suptitle('Gráfico de $θ(t)$ en funciondel tiempo',fontsize=20, fontweight='bold')
plt.ylabel('rad s')
plt.xlabel('Timepo (s)')
plt.legend(loc='best')
plt.show()
```
